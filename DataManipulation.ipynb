```python
import urllib.request
import json
import pprint

Url = 'https://prodapi.metweb.ie/monthly-data/Claremorris'

# Retrieve the JSON
response = urllib.request.urlopen(Url)
raw_json = response.read()

# Save JSON data to local file
f = open("content\WeatherClaremorris.json", 'wb')
f.write( raw_json )
f.close()

# Parse the JSON
WeatherClaremorris = json.loads(raw_json)

# Display the data
import pprint
pprint.pprint(WeatherClaremorris)
```

```python
#Display types of data
print(WeatherClaremorris.keys())
```

```python
print(len(WeatherClaremorris["mean_temperature"]["report"]))
```

```python
# Display county name of all items in the list
print(WeatherClaremorris["mean_temperature"])
for item in WeatherClaremorris["mean_temperature"]["report"]:
    print(item)
```

```python
import pprint
pprint.pprint(WeatherClaremorris["mean_temperature"]["report"])
```

```python
d = WeatherClaremorris['mean_temperature']["report"]
temp1 = []
for x in d.keys():
  for y in d[x].keys():
    #print(x,y + " " + d[x][y])
    try:
      temp1.append([x, y, float(d[x][y])])
    except:
      temp1.append([x, y, 0])
    #temp.append([x,y, d[x][y]])
    
df = pd.DataFrame(temp1, columns = ['Year', 'Month', 'Temperature'])

df
```

```python
df2 =df.groupby('Year').mean('Temperature')
df2
```

```python
d = WeatherClaremorris['total_rainfall']["report"]
rainfall = []
for x in d.keys():
  for y in d[x].keys():
    #print(x,y + " " + d[x][y])
    try:
      rainfall.append([x, y, float(d[x][y])])
    except:
      rainfall.append([x, y, 0])
    #temp.append([x,y, d[x][y]])
    
df_rain = pd.DataFrame(rainfall, columns = ['Year', 'Month', 'rainfall'])

df_rain
```

```python
d = WeatherClaremorris['evaporation']["report"]
evap = []
for x in d.keys():
  for y in d[x].keys():
    #print(x,y + " " + d[x][y])
    try:
      evap.append([x, y, float(d[x][y])])
    except:
      evap.append([x, y, 0])
    #temp.append([x,y, d[x][y]])
    
df_evap = pd.DataFrame(evap, columns = ['Year', 'Month', 'evaporation'])

df_evap
```

```python
All_data = pd.merge(df, df_rain, on = ['Year', 'Month'])
```

```python
All_data2 = pd.merge(All_data, df_evap, on = ['Year', 'Month'])
```
